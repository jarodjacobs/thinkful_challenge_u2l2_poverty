Import needed packages

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

Load csv file

```python
poverty_16 = pd.read_csv('/Users/jAir/jarod347/poverty_16.csv', index_col=1)
```

View data to make sure it loaded as expected

```python
poverty_16
```

List columns to help with selecting the ones that are most relevant

```python
list(poverty_16)
```

Select relevant columns

```python
poverty_sub = poverty_16.loc[:, ['Total Population','All Ages', 'Age 0-17',  
                                 'Age 5-17 in Families',
                                 'Age 0-4', 'Median Household Income', 'Welfare']]
```

Select rows 1-51 (excluding row for US). View dataframe to make sure the correct columns are selected.

```python
poverty_sub = poverty_sub.iloc[1:51, :]
poverty_sub
```

Visualize data with a bar chart to begin basic data exploration.

```python
poverty_sub.iloc[1:51, 1:].plot(kind = 'bar')
plt.show()
sns.set()
```

Add new column to dataframe that contains the percent of people in poverty which allows for a better comparison between levels of poverty in each state.

```python
poverty_sub['perc_pop_pov'] = poverty_sub['All Ages']/poverty_sub['Total Population']
```

View dataframe to make sure the column was added correctly.

```python
poverty_sub
```

Create a bar chart of the percentage of the population of each state living in poverty

```python
poverty_sub.iloc[0:-1, -1].plot(figsize = (15,6), kind = 'bar')
plt.show()
```

Create a new column that contains the number of welfare dollars spent for each person in that state. Multiply the Welfare column aby 1 billion as that column's data is in the billions).

```python
poverty_sub['welfare_dollars_per_person'] = (poverty_sub['Welfare']*1000000000)/poverty_sub['Total Population']
poverty_sort = poverty_sub.sort_values(by=['perc_pop_pov'])
```

Visualize dataframe to make sure the new column was added correctly.

```python
poverty_sort
```

Create a bar chart showing the number of welfare dollars spent per person in each state.

```python
poverty_sort.iloc[0:, -1].plot(figsize = (15,6), kind = 'bar')
plt.show()
```

Create three new columns: 1) z_score_perc_pov contains the z-score of perc_pop_pov; 2) perc_pop_child contains the percentage of the population that are children living in poverty; 3) z_score_perc_child contains the z_scores for perc_pop_child. The z-scores are calculated to help understand better how each state's poverty rates compare.

```python
poverty_sub['z_score_perc_pov'] = (poverty_sub.perc_pop_pov - poverty_sub.perc_pop_pov.mean())/poverty_sub.perc_pop_pov.std(ddof=0)
poverty_sub['perc_pop_child'] = poverty_sub['Age 0-4']/poverty_sub['Total Population']
poverty_sub['z_score_perc_child'] = (poverty_sub.perc_pop_child - poverty_sub.perc_pop_child.mean())/poverty_sub.perc_pop_child.std(ddof=0)
```

Sort the data by z-score to help visualize the difference between each state.

Plot a bar chart of the z-scores for the percentage of people living in poverty for each state. Add colors to help emphasize the states with low poverty rates (in blue) and those with high poverty rates (in red)

```python
poverty_sort_z = poverty_sub.sort_values(by=['z_score_perc_pov'])

colors = plt.cm.coolwarm(np.linspace(0,1,51))
poverty_sort_z.iloc[0:, -3].plot(figsize = (15,6), kind = 'bar', stacked=True, color = colors)
plt.show()
```

Produce a similar bar chart but with the data for children (0-4) living in poverty. Keep data sorted by z-score for total population living in poverty to help highlight the differences. Bars that are "out of order" show states that have higher/lower rates of children in poverty than the overall population.

```python
poverty_sort_z.iloc[0:, -3].plot(figsize = (15,6), kind = 'bar', stacked=True, color = colors)
poverty_sort_z.iloc[0:, -1].plot(figsize = (15,6), kind = 'bar', stacked=True, color = colors)
plt.show()
```

Use subplots to compare total poverty rates to the percentage of children living in poverty (here I'm focusing on children between the ages of 0 and 4).

```python
fig, axes = plt.subplots(nrows=2, ncols=1)
poverty_sort_z.iloc[0:, -3].plot(figsize = (15,6), kind = 'bar', stacked=True, color = colors, ax=axes[0])
poverty_sort_z.iloc[0:, -1].plot(figsize = (15,6), kind = 'bar', stacked=True, color = colors, ax=axes[1])
plt.show()
```
